## Creating Model

```python
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

## Spliting the Dataframe

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=48,test_size=0.20)
```

```python
svc=SVC(kernel='linear')
svc.fit(x_train,y_train)
y_pred=svc.predict(x_test)
print("This is accuracy_score", accuracy_score(y_pred,y_test))
print("This is confusion_matrix",confusion_matrix(y_pred,y_test)) 
print("This is the classification_report",classification_report(y_pred,y_test))
```

## KNN (K_Nearest_Neighbors_Classifier)

```python
kn=KNeighborsClassifier(n_neighbors=21)
kn.fit(x_train,y_train)
kn_pred=kn.predict(x_test)
print("This is accuracy_score", accuracy_score(kn_pred,y_test))
print("This is confusion_matrix",confusion_matrix(kn_pred,y_test)) 
print("This is the classification_report",classification_report(kn_pred,y_test))
```

## Tuning the Model

```python
pred_kn=kn.predict(x_test)
from sklearn.model_selection import cross_val_score
kn_score=accuracy_score(y_test,pred_kn)
for i in range(2,11):
    knscore=cross_val_score(kn,x,y,cv=i)
    knc=knscore.mean()
    print('at cv', i)
    print('cross_validation_score is:',knc*100)
    print('Accuracy_score is',kn_score*100)
    print("\n")
```
