## Data Cleansing

```python
df.corr()['Happiness Score']
```

### Dropping the negatively correlated and highly skewed data columns

```python
df.drop(['Standard Error'],axis=1,inplace=True)
df.head()
```

## Data Encoding

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
df['Country']=le.fit_transform(df['Country'])
df['Region']=le.fit_transform(df['Region'])
```

### Removing Outliers:

```python
from scipy.stats import zscore
import numpy as np
z=np.abs(zscore(df))
z.shape
```

```python
threshold=3
print(np.where(z>3))
```

```python
len(np.where(z>3)[0])
```

```python
z.iloc[155]
```

Dystopia Residual has maximum number of outtliers

```python
df_new=df[(z<3).all(axis=1)]
print("old DataFrame",df.shape)
print("New DataFrame",df_new.shape)
print("total_dropped_rows",df.shape[0]-df_new.shape[0])
```

```python
loss_percent=(158-152)/158*100
print(loss_percent,"%")
```

```python
x=df_new.drop(["Happiness Score"],axis=1)
y=df_new.iloc[:,3]
```

## Removing Skewness

```python
from sklearn.preprocessing import power_transform
x=power_transform(x,method='yeo-johnson')
x
```

```python
pd.DataFrame(x).skew()
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x=sc.fit_transform(x)
x
```

## Creating a best fit ML Model for prediction

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.ensemble import AdaBoostRegressor
```

## Spliting the Dataframe

## creating and training the model


```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=49,test_size=0.20)
lr= LinearRegression()
lr.fit(x_train,y_train)
pred=lr.predict(x_test)
print("result from Linear Regresssion Model")
print("R2 score is",r2_score(pred,y_test))
print("mean_squared_error",mean_squared_error(pred,y_test))
```

```python
x_train.shape
```

```python
x_test.shape
```

```python
y_train.shape
```

```python
y_test.shape
```

Thus our model has an accuracy of 99%

```python
rf=RandomForestRegressor(n_estimators=1000)
rf.fit(x_train,y_train)
pred=rf.predict(x_test)
print('result from random_forestregressor')
print('R2 score is',r2_score(pred,y_test))
print('mean_squared_error',mean_squared_error(pred,y_test))
```

```python
ab= AdaBoostRegressor()
ab.fit(x_train,y_train)
pred=ab.predict(x_test)
print('Result from AdaBoostRegressor')
print('R2 score is',r2_score(pred,y_test))
print('mean_squared_error',mean_squared_error(pred,y_test))
```

## Model_scores of linear regression, raandom forest,adaboost regressor

```python
reg=lr.score(x_train,y_train)
rand=rf.score(x_train,y_train)
ada=ab.score(x_train,y_train)

print(reg)
print(rand)
print(ada)
```

## Tuning the Model

### using k- Fold cross-validation technique

```python
pred_rf=rf.predict(x_test)
from sklearn.model_selection import cross_val_score
lrf=r2_score(y_test,pred_rf)
for i in range(2,10):
    lrfscore=cross_val_score(rf,x,y,cv=i)
    lrfs=lrfscore.mean()
    print('at cv', i)
    print('cross_validation_score is:',lrfs*100)
    print('R2 score is',lrf*100)
    print("\n")
```

### Selection of validation score from CV

```python
lrfscore_selected=cross_val_score(rf,x,y,cv=5).mean()
print('CV score is',lrfscore_selected,"\n accuracy_score is",lrfs)
```
