# Red Wine quality Prediction

# Problem Statement

Problem Statement:
The dataset is related to red and white variants of the Portuguese "Vinho Verde" wine. Due to privacy and logistic issues, only physicochemical (inputs) and sensory (the output) variables are available (e.g. there is no data about grape types, wine brand, wine selling price, etc.).

This dataset can be viewed as classification task. The classes are ordered and not balanced (e.g. there are many more normal wines than excellent or poor ones). Also, we are not sure if all input variables are relevant. So it could be interesting to test feature selection methods.

## Reducing thedimensions

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=4)
```

```python
pca.fit(x)
```

```python
x=pca.transform(x)
```

## Creating Model

```python
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix,accuracy_score,classification_report
```

## Spliting the Dataframe

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=48,test_size=0.20)
```

```python
svc=SVC(kernel='linear')
svc.fit(x_train,y_train)
y_pred=svc.predict(x_test)
print("This is accuracy_score", accuracy_score(y_pred,y_test))
print("This is confusion_matrix",confusion_matrix(y_pred,y_test)) 
print("This is the classification_report",classification_report(y_pred,y_test))
```

## KNN (K_Nearest_Neighbors_Classifier)

```python
kn=KNeighborsClassifier(n_neighbors=21)
kn.fit(x_train,y_train)
kn_pred=kn.predict(x_test)
print("This is accuracy_score", accuracy_score(kn_pred,y_test))
print("This is confusion_matrix",confusion_matrix(kn_pred,y_test)) 
print("This is the classification_report",classification_report(kn_pred,y_test))
```

## Tuning the Model

```python
pred_kn=kn.predict(x_test)
from sklearn.model_selection import cross_val_score
kn_score=accuracy_score(y_test,pred_kn)
for i in range(2,11):
    knscore=cross_val_score(kn,x,y,cv=i)
    knc=knscore.mean()
    print('at cv', i)
    print('cross_validation_score is:',knc*100)
    print('Accuracy_score is',kn_score*100)
    print("\n")
```

# Roc_Auc score for model

```python
from sklearn.metrics import roc_auc_score
```

```python
kn_pred_proba=kn.predict_proba(x_test)
```

```python
roc_auc_score(y_test,kn_pred_proba,multi_class='ovr')
```
